# Case study on Presentation of Car's data

# 1. Read the Car dataset to the python environment.

```python
import pandas as pd
cars_data=pd.read_csv("D:\python programming\cars_data.csv")
cars_data.head() ##This function returns the first 5 rows for the car dataset based on position
```

## 2. Check for the null values present in the car dataset

```python
## return the number of missing values in the car dataset
cars_data.isnull().sum()
```

## 3. Plot a bar graph of male vs female buyers participated in the sales

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.countplot(y='Make',hue='Buyer Gender',data=cars_data)
sns.set(rc={'figure.figsize':(50,50)})
plt.xticks(rotation=90)
plt.title('Gender wise participation in sales')
plt.legend(loc=1)
```

## 4. Find the top 5 cars based on their sales price

```python
cardata=cars_data.sort_values(by=['Sale Price'], ascending=False)
##Setting the ascending argument to False will sort the data the other way round, from large to small
cardata.head(5)
```

## 5. Find the least 5 cars based on their Resell price.

```python
cardata=cars_data.sort_values(by=['Resell Price'])
##sort the rows by passing a column name to .sort_values().
cardata.head(5)
```
